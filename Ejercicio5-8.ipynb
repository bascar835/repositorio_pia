```python
# Clase Persona
class Persona:
    def __init__(self, nombre, edad):
        self.nombre = nombre
        self.edad = edad

    def mostrar_info(self):
        print(f"Nombre: {self.nombre}")
        print(f"Edad: {self.edad}")

# Clase Empleado (hereda de Persona)
class Empleado(Persona):
    def __init__(self, nombre, edad, salario):
        super().__init__(nombre, edad)  # Llama al constructor de Persona
        self.salario = salario

    def mostrar_info(self):
        super().mostrar_info()  # Llama a mostrar_info de Persona
        print(f"Salario: {self.salario}")

# Clase Directivo (hereda de Empleado)
class Directivo(Empleado):
    def __init__(self, nombre, edad, salario, departamento):
        super().__init__(nombre, edad, salario)  # Llama al constructor de Empleado
        self.departamento = departamento

    def mostrar_info(self):
        super().mostrar_info()  # Llama a mostrar_info de Empleado
        print(f"Departamento: {self.departamento}")

    def asignar_departamento(self, nuevo_departamento):
        self.departamento = nuevo_departamento
        print(f"El departamento de {self.nombre} ha sido cambiado a {self.departamento}.")

# Crear una persona
persona = Persona("Juan Pérez", 30)
persona.mostrar_info()

print("\n---\n")

# Crear un empleado
empleado = Empleado("Ana González", 25, 3000)
empleado.mostrar_info()

print("\n---\n")

# Crear un directivo
directivo = Directivo("Carlos Méndez", 40, 8000, "Marketing")
directivo.mostrar_info()

print("\n---\n")

# Cambiar el departamento de un directivo
directivo.asignar_departamento("Ventas")

print("\n---\n")
directivo.mostrar_info()
```

```python
class Contacto:
    def __init__(self, nombre, telefono):
        self.nombre = nombre
        self.telefono = telefono
    
    def actualizar_telefono(self, nuevo_telefono):
        self.telefono = nuevo_telefono
        print(f"El teléfono de {self.nombre} ha sido actualizado.")

class Agenda:
    def __init__(self):
        self.contactos = []

    def agregar_contacto(self, nombre, telefono):
        contacto = Contacto(nombre, telefono)
        self.contactos.append(contacto)
        print(f"Contacto de {nombre} con teléfono {telefono} agregado exitosamente.")
    
    def eliminar_contacto(self, nombre):
        for contacto in self.contactos:
            if contacto.nombre.lower() == nombre.lower():
                self.contactos.remove(contacto)
                print(f"Contacto de {nombre} eliminado.")
                return
        print(f"Contacto con el nombre {nombre} no encontrado.")
    
    def actualizar_telefono_contacto(self, nombre, nuevo_telefono):
        for contacto in self.contactos:
            if contacto.nombre.lower() == nombre.lower():
                contacto.actualizar_telefono(nuevo_telefono)
                return
        print(f"Contacto con el nombre {nombre} no encontrado.")
    
    def buscar_contacto(self, nombre):
        for contacto in self.contactos:
            if contacto.nombre.lower() == nombre.lower():
                print(f"Nombre: {contacto.nombre}, Teléfono: {contacto.telefono}")
                return
        print(f"Contacto con el nombre {nombre} no encontrado.")
    
    def mostrar_contactos(self):
        if not self.contactos:
            print("No hay contactos en la agenda.")
        else:
            for contacto in self.contactos:
                print(f"Nombre: {contacto.nombre}, Teléfono: {contacto.telefono}")

# Función que muestra el menú interactivo
def menu(agenda):
    while True:
        print("\nMenú de la Agenda de Contactos:")
        print("1. Agregar contacto")
        print("2. Eliminar contacto")
        print("3. Actualizar teléfono de contacto")
        print("4. Buscar contacto")
        print("5. Mostrar todos los contactos")
        print("6. Salir")
        
        opcion = input("Elige una opción: ")

        if opcion == "1":
            nombre = input("Ingresa el nombre del contacto: ")
            telefono = input("Ingresa el número de teléfono: ")
            agenda.agregar_contacto(nombre, telefono)
        
        elif opcion == "2":
            nombre = input("Ingresa el nombre del contacto a eliminar: ")
            agenda.eliminar_contacto(nombre)
        
        elif opcion == "3":
            nombre = input("Ingresa el nombre del contacto a actualizar: ")
            nuevo_telefono = input("Ingresa el nuevo número de teléfono: ")
            agenda.actualizar_telefono_contacto(nombre, nuevo_telefono)
        
        elif opcion == "4":
            nombre = input("Ingresa el nombre del contacto a buscar: ")
            agenda.buscar_contacto(nombre)
        
        elif opcion == "5":
            agenda.mostrar_contactos()
        
        elif opcion == "6":
            print("Saliendo de la agenda...")
            break
        
        else:
            print("Opción no válida. Intenta de nuevo.")

# Bloque de ejecución principal
if __name__ == "__main__":
    agenda = Agenda()
    menu(agenda)

```

```python
from abc import ABC, abstractmethod

# Clase abstracta base Animal
class Animal(ABC):
    def __init__(self, nombre, edad):
        self.nombre = nombre
        self.edad = edad
    
    @abstractmethod
    def hacer_sonido(self):
        pass
    
    @abstractmethod
    def moverse(self):
        pass


# Clase Mamífero
class Mamifero(Animal):
    def hacer_sonido(self):
        return f"El {self.nombre} hace un gruñido."
    
    def moverse(self):
        return f"El {self.nombre} camina."


# Clase Ave
class Ave(Animal):
    def hacer_sonido(self):
        return f"El {self.nombre} canta."
    
    def moverse(self):
        return f"El {self.nombre} vuela."


# Clase Reptil
class Reptil(Animal):
    def hacer_sonido(self):
        return f"El {self.nombre} hace un susurro."
    
    def moverse(self):
        return f"El {self.nombre} se arrastra."


# Clase Zoologico
class Zoologico:
    def __init__(self):
        self.animales = []

    def agregar_animal(self, tipo, nombre, edad):
        if tipo.lower() == 'mamifero':
            self.animales.append(Mamifero(nombre, edad))
        elif tipo.lower() == 'ave':
            self.animales.append(Ave(nombre, edad))
        elif tipo.lower() == 'reptil':
            self.animales.append(Reptil(nombre, edad))
        else:
            print("Tipo de animal no reconocido.")
    
    def mostrar_animales(self):
        for animal in self.animales:
            print(f"{animal.__class__.__name__} - Nombre: {animal.nombre}, Edad: {animal.edad}")
    
    def hacer_sonidos(self):
        for animal in self.animales:
            print(animal.hacer_sonido())
    
    def mover_animales(self):
        for animal in self.animales:
            print(animal.moverse())


# Menú interactivo
def menu(zoologico):
    opciones = {
        "1": zoologico.agregar_animal,
        "2": zoologico.mostrar_animales,
        "3": zoologico.hacer_sonidos,
        "4": zoologico.mover_animales,
        "5": exit
    }
    
    while True:
        print("\n--- Menú Zoológico ---")
        print("1. Agregar un animal")
        print("2. Mostrar todos los animales")
        print("3. Hacer que todos los animales hagan sonido")
        print("4. Hacer que todos los animales se muevan")
        print("5. Salir")
        
        opcion = input("Selecciona una opción: ")

        match opcion:
            case "1":
                tipo = input("Tipo de animal (mamífero, ave, reptil): ")
                nombre = input("Nombre del animal: ")
                edad = int(input("Edad del animal: "))
                opciones[opcion](tipo, nombre, edad)
            case "2":
                opciones[opcion]()
            case "3":
                opciones[opcion]()
            case "4":
                opciones[opcion]()
            case "5":
                print("Saliendo del programa.")
                break
            case _:
                print("Opción no válida. Intenta nuevamente.")


# Estructura principal
if __name__ == "__main__":
    zoologico = Zoologico()
    menu(zoologico)
```

```python
from abc import ABC, abstractmethod

# Clase base abstracta para todos los animales
class Animal(ABC):
    def __init__(self, nombre, edad):
        if not self.validar_nombre(nombre):
            raise ValueError("El nombre del animal no es válido.")
        self.nombre = nombre
        self.edad = edad

    @abstractmethod
    def hacer_sonido(self):
        pass

    @abstractmethod
    def moverse(self):
        pass

    @staticmethod
    def validar_nombre(nombre):
        # Validar que el nombre no esté vacío y que no contenga caracteres no válidos
        return bool(nombre) and nombre.isalpha()

# Clases de habilidades especiales
class Volador:
    def volar(self):
        return f"{self.nombre} está volando."

class Acuatico:
    def nadar(self):
        return f"{self.nombre} está nadando."

# Clases hijas específicas de Animal
class Mamifero(Animal):
    def hacer_sonido(self):
        return f"{self.nombre} hace un sonido de mamífero."

    def moverse(self):
        return f"{self.nombre} está caminando."

class Ave(Animal, Volador):
    def hacer_sonido(self):
        return f"{self.nombre} hace un sonido de ave."

    def moverse(self):
        return self.volar()

class Reptil(Animal):
    def hacer_sonido(self):
        return f"{self.nombre} hace un sonido de reptil."

    def moverse(self):
        return f"{self.nombre} se está arrastrando."

# Ejemplo de herencia múltiple para el pato
class Pato(Animal, Volador, Acuatico):
    def hacer_sonido(self):
        return f"{self.nombre} hace cuac cuac."

    def moverse(self):
        return f"{self.nombre} puede caminar, {self.volar()} y {self.nadar()}"

# Clase Habitat para gestionar hábitats en el zoológico
class Habitat:
    def __init__(self, nombre, capacidad_maxima):
        self.nombre = nombre
        self.capacidad_maxima = capacidad_maxima
        self.animales = []

    def agregar_animal(self, animal):
        if len(self.animales) < self.capacidad_maxima:
            self.animales.append(animal)
            print(f"{animal.nombre} fue añadido al hábitat {self.nombre}.")
        else:
            print(f"El hábitat {self.nombre} está lleno.")

    def mostrar_animales(self):
        for animal in self.animales:
            print(f"{animal.nombre}, Tipo: {type(animal).__name__}, Edad: {animal.edad}")

    def contar_animales(self):
        return len(self.animales)

    def __add__(self, otro_habitat):
        return self.contar_animales() + otro_habitat.contar_animales()

# Clase Zoologico para gestionar los hábitats y animales
class Zoologico:
    def __init__(self):
        self.habitats = []

    def agregar_habitat(self, nombre, capacidad_maxima):
        habitat = Habitat(nombre, capacidad_maxima)
        self.habitats.append(habitat)
        print(f"Hábitat {nombre} con capacidad para {capacidad_maxima} animales añadido al zoológico.")

    def agregar_animal(self, tipo, nombre, edad, nombre_habitat):
        try:
            habitat = next((h for h in self.habitats if h.nombre == nombre_habitat), None)
            if habitat:
                if tipo == "mamífero":
                    animal = Mamifero(nombre, edad)
                elif tipo == "ave":
                    animal = Ave(nombre, edad)
                elif tipo == "reptil":
                    animal = Reptil(nombre, edad)
                elif tipo == "pato":
                    animal = Pato(nombre, edad)
                else:
                    print("Tipo de animal no reconocido.")
                    return
                habitat.agregar_animal(animal)
            else:
                print(f"No se encontró el hábitat con nombre {nombre_habitat}.")
        except ValueError as e:
            print(f"Error al agregar el animal: {e}")

    def mostrar_habitats(self):
        for habitat in self.habitats:
            print(f"Hábitat: {habitat.nombre}, Capacidad: {habitat.capacidad_maxima}, Animales: {habitat.contar_animales()}")
            habitat.mostrar_animales()

    def contar_total_animales(self):
        return sum(habitat.contar_animales() for habitat in self.habitats)

    def sumar_habitats(self, nombre_habitat1, nombre_habitat2):
        habitat1 = next((h for h in self.habitats if h.nombre == nombre_habitat1), None)
        habitat2 = next((h for h in self.habitats if h.nombre == nombre_habitat2), None)
        if habitat1 and habitat2:
            return habitat1 + habitat2
        else:
            print("Uno o ambos hábitats no fueron encontrados.")
            return None

# Menú interactivo
def main():
    zoologico = Zoologico()
    opciones = {
        1: "Agregar un hábitat",
        2: "Agregar un animal a un hábitat",
        3: "Mostrar todos los hábitats y animales",
        4: "Contar el total de animales en el zoológico",
        5: "Sumar animales entre dos hábitats",
        6: "Salir"
    }

    while True:
        print("\nOpciones:")
        for k, v in opciones.items():
            print(f"{k}. {v}")
        
        try:
            opcion = int(input("Seleccione una opción: "))
            if 1 <= opcion <= 6:
                if opcion == 1:
                    nombre = input("Nombre del hábitat: ")
                    capacidad = int(input("Capacidad máxima del hábitat: "))
                    zoologico.agregar_habitat(nombre, capacidad)
                elif opcion == 2:
                    tipo = input("Tipo de animal (mamífero, ave, reptil, pato): ").lower()
                    nombre = input("Nombre del animal: ")
                    edad = int(input("Edad del animal: "))
                    nombre_habitat = input("Nombre del hábitat: ")
                    zoologico.agregar_animal(tipo, nombre, edad, nombre_habitat)
                elif opcion == 3:
                    zoologico.mostrar_habitats()
                elif opcion == 4:
                    total_animales = zoologico.contar_total_animales()
                    print(f"Total de animales en el zoológico: {total_animales}")
                elif opcion == 5:
                    nombre_habitat1 = input("Nombre del primer hábitat: ")
                    nombre_habitat2 = input("Nombre del segundo hábitat: ")
                    total = zoologico.sumar_habitats(nombre_habitat1, nombre_habitat2)
                    if total is not None:
                        print(f"Total de animales en ambos hábitats: {total}")
                elif opcion == 6:
                    print("Saliendo del programa...")
                    break
            else:
                print("Por favor, ingrese un número entre 1 y 6.")
        except ValueError:
            print("Por favor, ingrese un número válido.")

# Ejecuta el menú interactivo
main()
```
